```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.model_selection import cross_val_score
```

```python
import sqlite3
import os
```

```python
connection = sqlite3.connect("./reviews_crawler/reviews.db")
cursor = connection.cursor()
```

```python
cursor.execute("SELECT _id, rating FROM reviews")
result = cursor.fetchall()
```

```python
result
```

```python
texts = []
ratings = []
for item in result:
    if not item[1] is None:
        if 0 <= int(item[1]) <= 40:
            ratings.append('flop')
        elif 41 <= int(item[1]) <= 60:
            ratings.append('meh')
        elif 61 <= int(item[1]) <= 100:
            ratings.append('top')
        with open("./reviews_crawler/review_texts/" + item[0]) as review:
            texts.append(review.readline())
```

```python
vectorizer = TfidfVectorizer(ngram_range=(1, 3))
classifier = LogisticRegression()

model = Pipeline([
    ('vectorizer', vectorizer),
    ('classifier', classifier)
])

scores = cross_val_score(model, texts, ratings, cv=5, scoring='f1_weighted')
```

```python
import numpy as np
np.average(scores)
```
